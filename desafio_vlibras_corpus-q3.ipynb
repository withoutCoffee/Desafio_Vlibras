##  Script para fazer um data augmentation

```python
import re
import pandas as pd
```

```python
data =  pd.read_csv('corpora/corpus-q3.csv')
```

### Função que gera todas as combinações possíveis para os verbos

```python
from itertools import product,combinations

nums = ['1','2','3']
lts = ['S','P']        

```

```python
def combination_verbs(verb):
    comb = product(nums,lts, repeat=1)
    comb = product(comb, repeat=2)
    
    result = []
    for i in comb:

        a = i[0][0]+i[0][1]#lado esqueda do verbo
        b = i[1][0]+i[1][1]#lado esqueda do verbo
        
        result.append(a+verb+b)
    return result
```

### Criando novas frases

```python
result_all = []

for i in range(len(data)):
    x = re.findall('[1-3][SP]\w+[1-3][SP]',data['gi'][i])
    
    left_verb = re.findall('(.+)[1-3][SP]\w+[1-3][SP]',data['gi'][i])#pegando o lago esquerdo do verbo
    
    right_verb = re.findall('[1-3][SP]\w+[1-3][SP](.+)',data['gi'][i])#pegando o lago direito do verbo
    
    #print(left_verb[0],'ººº',right_verb[0])
    
    if x != []:    
        y = re.findall('_\w+_',x[0])
        y = y[0]
        
        
        for j in combination_verbs(y):
            
            result_all.append(left_verb[0]+j+right_verb[0])
```

```python
result_all
```

### Escrevendo novas frases em arquivo txt

```python
file = open('corpus-respostas/corpus-q3-resposta.txt','w')

for i in result_all:
    file.write(i+'\n')
```

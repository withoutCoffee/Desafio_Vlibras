## Frequências para as palavras presentes no arquivo “corpus-q2.csv”

```python
import re
import pandas as pd
import json
```

```python
data = pd.read_csv('corpora/corpus-q2.csv')
```

```python
data.head()
```

```python
hashtable = {}

for i in range(len(data)):
    
    x = re.findall('\w+',data['gr'][i])#para coluna GR   
    
    y = re.findall('\w+',data['gi'][i])#para coluna GI   
    
    #para coluna GR   
    for j in range(len(x)):
        
        word = x[j].lower()
        if word in hashtable:
            
            hashtable[word] = hashtable[word] + 1
            
        else:
            hashtable[word] = 1
            
    #para coluna GI   
    for j in range(len(y)):
        
        word = y[j].lower()
        if word in hashtable:
            
            hashtable[word] = hashtable[word] + 1
            
        else:
            hashtable[word] = 1
```
